```python
import pandas as pd
import numpy as np
```

## Read Tables

```python
df1 = pd.read_csv("iverson.test.csv",index_col=0)
df2 = pd.read_csv("mcgrady.test.csv",index_col=0)
```

```python
df1.head()
```

```python
df2.head()
```

## Merge Tables

### Inner Join

```python
df1.merge(df2,how="inner",left_on="Season",right_on="Season")
```

### Outer Join

```python
df1.merge(df2,how="outer",left_on="Season",right_on="Season")
```

### Left Join

```python
df1.merge(df2,how="left",left_on="Season",right_on="Season")
```

### Right Join

```python
df1.merge(df2,how="right",left_on="Season",right_on="Season")
```

### Merge On Multiple Columns

```python
df1.merge(df2,how="inner",left_on=["Season","Pos"],right_on=["Season","Pos"])
```

### Merge Multiple Data Frames

```python
df3 = pd.read_csv("kobe.test.csv",index_col=0)
data_frames = [df1, df2, df3]
from functools import reduce
reduce(lambda x,y: x.merge(y,left_on='Season',right_on="Season",how='inner'), data_frames)
```
